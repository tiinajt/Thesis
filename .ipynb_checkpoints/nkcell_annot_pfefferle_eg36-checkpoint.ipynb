## T and NK cell markers by Van Galen et al. 

```python
t_ctl=["CD3D", "CD3G"] 
t_ctl_nk = ["IL32", "IL7R"]
t_cell = ["TCF7"]
ctl_nk = ["CCL5"]
ctl = ["GZMK"]
nk =[ "KLRB1", "KLRD1", "GZMB", "NCAM1"]
```

```python
listOfGenes=t_ctl

t_ctl_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        t_ctl_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
listOfGenes=t_ctl_nk

t_ctl_nk_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        t_ctl_nk_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
listOfGenes=t_cell

t_cell_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        t_cell_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
listOfGenes=ctl_nk
ctl_nk_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        ctl_nk_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
listOfGenes=ctl
ctl_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        ctl_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
listOfGenes=nk 
nk_vangalen=[]

for x in listOfGenes:
    if x in adata1.var_names:
        print(x, "Yes")
        nk_vangalen.append(x)
    else:
        print(x, "No") 
        
print(listOfGenes)
```

```python
vangalen_markers= [t_ctl_vangalen, t_ctl_nk_vangalen, t_cell_vangalen, ctl_nk_vangalen, ctl_vangalen, nk_vangalen] 
```

```python
vangalen_markers = list(np.concatenate(vangalen_markers).flat)

vangalen_markers
```

```python
sc.settings.set_figure_params(dpi=300)
```

```python
sc.pl.umap(adata1, color = vangalen_markers, color_map = 'viridis', ncols = 1)
```

```python
g = sc.pl.violin(adata, keys= vangalen_markers, groupby= 'cell_type', scale = 'counts')
```
