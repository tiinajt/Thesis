```python
#Selecting only easigenomics
adata_easi=adata[adata.obs['batch_group'] == 'Easi' ].copy()
```

### Adding DSB

```python
#List of sample names

eg_list=adata_dsb.obs["EG_code"].unique().tolist()
```

```python
result = None

for eg in eg_list:
    try:
        tail = adata.obs[adata.obs["eg_code"] == eg].index[0][-4:]
        new_index=adata_dsb.obs[adata_dsb.obs["EG_code"] == eg].index+tail
        if result is None:
            result=new_index.tolist()
        else:
            result+=(new_index.tolist())
    except:
        new_index=adata_dsb.obs[adata_dsb.obs["EG_code"] == eg].index
        result+=(new_index.tolist())
        
```

```python
len(result)
```

```python
adata_dsb.shape
```

```python
adata_dsb.obs.index=pd.Index(result)
```
